```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Iris.csv")
```

**DATA CLEANING**

```python
'''CHANGING THE SPECIES NAME
Iris-setosa -> Setosa
Iris-versicolor -> Versicolor
Iris-virginica -> Virginica'''
df["Species"]=df["Species"].replace("Iris-setosa","Setosa")
df["Species"]=df["Species"].replace("Iris-versicolor","Versicolor")
df["Species"]=df["Species"].replace("Iris-virginica","Virginica")
df["Species"].unique()
```

<br>

```python
# FINDING THE MEAN OF VERSICOLOR SEPAL LENGTH
df[df.Species=="Versicolor"]["SepalLengthCm"].mean()
```

```python
# FINDING THE MEDIAN OF VERSICOLOR SEPAL LENGTH
df[df.Species=="Versicolor"]["SepalLengthCm"].median()
```

**DATA VISUALIZATION**

*BAR CHART*

```python
# SEPAL WIDTH VISUALIZATION FOR ALL THE THREE SPECIES

LABEL=["Setosa","Versicolor","Virginica"]
s=df[df.Species=="Setosa"]["SepalWidthCm"].mean()
ve=df[df.Species=="Versicolor"]["SepalWidthCm"].mean()
vi=df[df.Species=="Virginica"]["SepalWidthCm"].mean()
VALUES=[s,ve,vi]
plt.bar(LABEL,VALUES,color="violet",width=0.5)
plt.ylabel("SEPAL WIDTH (CM)",labelpad=10)
plt.xlabel("SPECIES",labelpad=10)
plt.show()
```

```python
# SEPAL LENGTH VISUALIZATION FOR ALL THE THREE SPECIES

LABEL=["Setosa","Versicolor","Virginica"]
s=df[df.Species=="Setosa"]["SepalLengthCm"].mean()
ve=df[df.Species=="Versicolor"]["SepalLengthCm"].mean()
vi=df[df.Species=="Virginica"]["SepalLengthCm"].mean()
VALUES=[s,ve,vi]
plt.bar(LABEL,VALUES,color="lightblue",width=0.5)
plt.ylabel("SEPAL LENGTH (CM)",labelpad=10)
plt.xlabel("SPECIES",labelpad=10)
plt.show()
```

```python
# PETAL LENGTH VISUALIZATION FOR ALL THE THREE SPECIES

LABEL=["Setosa","Versicolor","Virginica"]
s=df[df.Species=="Setosa"]["PetalLengthCm"].mean()
ve=df[df.Species=="Versicolor"]["PetalLengthCm"].mean()
vi=df[df.Species=="Virginica"]["PetalLengthCm"].mean()
VALUES=[s,ve,vi]
plt.bar(LABEL,VALUES,color="hotpink",width=0.5)
plt.ylabel("PETAL LENGTH (CM)",labelpad=10)
plt.xlabel("SPECIES",labelpad=10)
plt.show()
```

```python
# PETAL WIDTH VISUALIZATION FOR ALL THE THREE SPECIES

LABEL=["Setosa","Versicolor","Virginica"]
s=df[df.Species=="Setosa"]["PetalWidthCm"].mean()
ve=df[df.Species=="Versicolor"]["PetalWidthCm"].mean()
vi=df[df.Species=="Virginica"]["PetalWidthCm"].mean()
VALUES=[s,ve,vi]
plt.bar(LABEL,VALUES,color="lightgreen",width=0.5)
plt.ylabel("PETAL WIDTH (CM)",labelpad=10)
plt.xlabel("SPECIES",labelpad=10)
plt.show()
```

*LINE CHART*

```python
# SEPAL LENGTH LINE GRAPH

plt.plot(df[df.Species=="Setosa"]["SepalLengthCm"])
plt.plot(df[df.Species=="Versicolor"]["SepalLengthCm"])
plt.plot(df[df.Species=="Virginica"]["SepalLengthCm"])
plt.legend(["Setosa","Versicolor","Virginica"])
plt.show()
```

```python
# SEPAL WIDTH LINE GRAPH

plt.plot(df[df.Species=="Setosa"]["SepalWidthCm"])
plt.plot(df[df.Species=="Versicolor"]["SepalWidthCm"])
plt.plot(df[df.Species=="Virginica"]["SepalWidthCm"])
plt.legend(["Setosa","Versicolor","Virginica"])
plt.show()
```

*BAR PLOT*

```python
# COMBINED VISUALIZATION OF PETAL LENGTH AND PETAL WIDTH OF ALL THE SPECIES
iris_melted = pd.melt(df, id_vars='Species', value_vars=['PetalLengthCm', 'PetalWidthCm'], var_name='Measurement', value_name='Value')
sns.barplot(x='Species', y='Value', hue='Measurement', data=iris_melted,palette=['violet','lightgreen'])
plt.show()
```

*SCATTER PLOT*

```python
# COMPARISON OF SEPAL DIMENSIONS BETWEEN SETOSA AND VERSICOLOR SPECIES
x=df[df.Species=="Setosa"]["SepalLengthCm"]
y=df[df.Species=="Setosa"]["SepalWidthCm"]
plt.scatter(x,y,color='hotpink')
x=df[df.Species=="Versicolor"]["SepalLengthCm"]
y=df[df.Species=="Versicolor"]["SepalWidthCm"]
plt.scatter(x,y)
plt.show()
```
